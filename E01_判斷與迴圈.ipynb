# 題目:流程控制：判斷式和迴圈，不能使用容器
## 說明 ##


請試著輸入一數字n：<br>
並列印出n * n之*號空心方框<br>


## Input Format ##


數字n <br>


## Output Format ##


長寬為n的*號空心方塊(米符號前含含空白、方框內皆空白)<br>
注意每一排最後一個*後面沒有空白<br>


## Sample Input 1 ##



```
3

```



## Sample Output 1 ##



```

 * * *
 *   *
 * * *
 
```



## Sample Input 2 ##



```
5

```



## Sample Output 2 ##



```

 * * * * *
 *       *
 *       *
 *       *
 * * * * *

```




## Sample Input 3 ##



```
8

```



## Sample Output 3 ##



```

 * * * * * * * *
 *             *
 *             *
 *             *
 *             *
 *             *
 *             *
 * * * * * * * *
 
```




## Hint ##



```
1. 注意第一排*號前含空白。
2. 使用判斷式和迴圈，不需要用容器。


```
